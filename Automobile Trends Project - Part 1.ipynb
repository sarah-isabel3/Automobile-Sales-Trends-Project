```python
#Fuel Efficiency Analysis
#Task 3

#plt.figure(figsize=(12, 8))
df.boxplot(column=['City_Mileage_km_litre'])
plt.title("City Mileage Box Plot (Before Outlier Treatment)")
plt.ylabel("City Mileage (km/litre)")

print("The boxplot shows the presence of outliers in city mileage")
```

```python
#Find outliers in city mileage

Q1 = df['City_Mileage_km_litre'].quantile(0.25)
Q3 = df['City_Mileage_km_litre'].quantile(0.75)
IQR = Q3 - Q1

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

outliers = df[(df['City_Mileage_km_litre'] < lower_fence) | 
              (df['City_Mileage_km_litre'] > upper_fence)]

print("Number of outliers:", len(outliers))
```

```python
#Resolve the outliers

# Treat outliers using clip
df['City_Mileage_km_litre'] = df['City_Mileage_km_litre'].clip(lower=lower_fence, upper=upper_fence)

# Check remaining outliers (should be zero)
remaining_outliers = df[
    (df['City_Mileage_km_litre'] < lower_fence) |
    (df['City_Mileage_km_litre'] > upper_fence)
]
print("Remaining outliers after treatment:", len(remaining_outliers))

#Save a copy of the treated DataFrame
df_clipped = df.copy()

# Step 5: Visualize with boxplot
plt.figure(figsize=(8, 5))
df.boxplot(column=['City_Mileage_km_litre'])
plt.title("City Mileage Boxplot (After treating Outliers)")
plt.ylabel("City Mileage (km/litre)")
plt.grid(True)
plt.show()
```

```python
#Task 11

#Preprocess the columns in main dataframe

def convert_kg_to_lbs(series):
    series = series.replace("Not Available", np.nan)
    series = series.astype(str).str.replace("kg", "", regex=False).str.strip()
    series = pd.to_numeric(series, errors="coerce")
    return series * 2.2

df["Kerb_Weight_lbs"] = convert_kg_to_lbs(df["Kerb_Weight"])
df["Gross_Vehicle_Weight_lbs"] = convert_kg_to_lbs(df["Gross_Vehicle_Weight"])

df["Kerb_Weight_lbs"] = df["Kerb_Weight_lbs"].fillna(df["Kerb_Weight_lbs"].median())
df["Gross_Vehicle_Weight_lbs"] = df["Gross_Vehicle_Weight_lbs"].fillna(df["Gross_Vehicle_Weight_lbs"].median())

print("Converted Kerb Weight (first 5 rows):")
print(df[["Kerb_Weight", "Kerb_Weight_lbs"]].head())

print("\nConverted Gross Vehicle Weight (first 5 rows):")
print(df[["Gross_Vehicle_Weight", "Gross_Vehicle_Weight_lbs"]].head())

```

```python
# Replace the Kerb Weight and Gross Vehicle Weight columns
df["Kerb_Weight"] = df["Kerb_Weight_lbs"]
df["Gross_Vehicle_Weight"] = df["Gross_Vehicle_Weight_lbs"]

df.drop(columns=["Kerb_Weight_lbs", "Gross_Vehicle_Weight_lbs"], inplace=True)

df_final = df.copy()
```

```python
#Performance Comparison
#Task 12

performance_cols = ['Displacement', 'Cylinders', 'Valves_Per_Cylinder']

# Calculate mean, median, std, and coefficient of variation (CV = std/mean)
stats_df = df.groupby('Make')[performance_cols].agg(
    ['mean', 'median', 'std']
)

stats_df.columns = ['_'.join(col) for col in stats_df.columns]

# Add coefficient of variation (per column)
for col in performance_cols:
    stats_df[f'{col}_cv'] = stats_df[f'{col}_std'] / stats_df[f'{col}_mean']

pd.set_option("display.max_columns", None)
print(stats_df.head())

#Box Plot of Displacement per Manufacturer

import seaborn as sns

plt.figure(figsize=(14, 6))
sns.boxplot(x='Make', y='Displacement', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Displacement across Manufacturers')
plt.ylabel('Displacement (cc)')
plt.xlabel('Manufacturer')
plt.show()
```

```python
"""
Performance Comparison
Task 12

- Aston Martin & Bentley have very high displacement (~5000â€“5500 cc), with multiple cylinders (avg. ~10â€“12). Low CV for cylinders and valves showing very consistent engineering design.

- Audi & BMW has mid-to-high displacement (~2200â€“2700 cc) with 4â€“6 cylinders. Audi shows the highest variation in valves per cylinder (CV ~0.63), meaning different valve technologies across models.

- Bajaj has very small displacement (~216 cc) and fixed 1 valve/cylinder. Bajaj also has zero variation meaning highly standardized products.

- Mass-market brands like Maruti Suzuki, Tata, Hyundai cluster between 800â€“1600 cc which focuses on affordability and efficiency.

- Luxury/sports brands (Bentley, Ferrari, Lamborghini) have very wide displacement range, often exceeding 6000 cc.
"""
```
