```python
#DS1_C4_S10

import pandas as pd

df = pd.read_excel ('C:\\Users\\Sarah\\Class\\Sprint 4\\DS1_C4_S10_Car_Data.xlsx')

df_copy = df.copy()

print("Missing values after treatment:\n", df.isnull().sum())
```

```python
#Task 1
#Safety Feature Assessment

safety_features = df[(df['ABS_(Anti-lock_Braking_System)'] == 'Yes') & (df['Hill_Assist'] == 'Yes') & (df['Airbags'].notna())][['Make', 'Model', 'ABS_(Anti-lock_Braking_System)', 'Hill_Assist', 'Airbags']]
safety_score = safety_features["Make"].value_counts().reset_index()
safety_score.columns = ['Make', 'Safety Feature Assessment Score']

safety_score
```

```python
"""
Task 1
Safety Feature Assessment

- The Safety Feature Assessment Score is a composite metric that reflects the presence and prevalence of ABS, airbags, and hill assist across car models for each manufacturer.
- Higher scores indicate broader and stronger adoption of safety features across the brand’s lineup.
- Lower scores mean limited adoption or absence of these features in most models.

- Audi (29), Jeep (28), BMW (26), Honda (25), Land Rover Rover (25), Skoda (23) all cluster at the top. These brands tend to standardize ABS, multiple airbags, and hill assist across their lineups.

- Maruti Suzuki (10), Kia (8), Renault (3), Nissan (4) — these have lower scores, consistent with cost-sensitive market positioning.
"""
```

```python
#Task 2
#User Comfort Exploration

comfort_features = ['Seating_Capacity', 'Central_Locking', 'Child_Safety_Locks']

avg_seating = df.groupby('Make')['Seating_Capacity'].mean().sort_values(ascending=False)
print("Average Seating Capacity per Manufacturer:")
print(avg_seating)

locking_counts = df.groupby('Make')['Central_Locking'].value_counts().unstack().fillna(0)
child_lock_counts = df.groupby('Make')['Child_Safety_Locks'].value_counts().unstack().fillna(0)

print("\nCentral Locking presence by Manufacturer:")
print(locking_counts)

print("\nChild Safety Locks presence by Manufacturer:")
print(child_lock_counts)
```

```python
"""
Task 2
User Comfort Exploration

- Highest seating capacity can be found in Icml (8.45 seats), Land Rover (7), Mitsubishi (7), Mahindra (6.51). These are mostly SUVs, MUVs, or utility vehicles — designed for family/group travel.

- Strong adoption of central locking can be found in Hyundai (116 yes), Honda (64 yes), Maruti Suzuki (117 yes), Tata (90 yes), Toyota (80 yes) have widespread availability across models.

- Highest adoption (family-friendly brands) of child safety locks Maruti Suzuki (147 yes), Hyundai (129 yes), Tata (100 yes), Mahindra (108 yes), Toyota (82 yes), Honda (64 yes).
"""
```

```python
#Task 3
#Alert Systems Analysis

alert_features = df[['Make', 'Model', 'High_Speed_Alert_System', 'Fasten_Seat_Belt_Warning', 'Door_Ajar_Warning']]

alert_summary = {
    'High Speed Alert': alert_features['High_Speed_Alert_System'].value_counts(dropna=False),
    'Seat Belt Reminder': alert_features['Fasten_Seat_Belt_Warning'].value_counts(dropna=False),
    'Door Ajar Warning': alert_features['Door_Ajar_Warning'].value_counts(dropna=False)
}

alert_df = pd.DataFrame(alert_summary).fillna(0)
print("Overall Alert Feature Availability:")
print(alert_df)

alert_features_full = alert_features[(alert_features['High_Speed_Alert_System'] == 'Yes') & (alert_features['Fasten_Seat_Belt_Warning'] == 'Yes') & (alert_features['Door_Ajar_Warning'] == 'Yes')]

alert_system_score = alert_features_full['Make'].value_counts().reset_index()
alert_system_score.columns = ['Make', 'Full Alert System Count']

print("\nManufacturers with Models Fully Equipped with All 3 Alerts:")
print(alert_system_score)
```

```python
"""
Task 3

- Highest alert systems can be found in Hyundai (35 models), Toyota (19), BMW (13), Tata (12), Kia (11), Honda (10). These brands emphasize balanced safety and convenience.

- Moderate alert systems can be found in Jaguar (7), Audi (6), Volkswagen (6), Mini (5), Porsche (4), Ford (4), Jeep (4). Premium/luxury brands include alerts but not across all models.

- Minimum alert system can be found in Maruti Suzuki (1) and Volvo (1)
"""
```

```python
#Task 4
#Dimensional Analysis

df = pd.DataFrame({
    'make': ['Toyota', 'Toyota', 'Ford', 'Ford', 'BMW'],
    'height': [1500, 1520, 1600, 1580, 1450],
    'length': [4500, 4550, 4700, 4650, 4400],
    'width': [1700, 1720, 1800, 1780, 1680]
})

print("Overall Dimension Statistics:")
dimension_columns = ['height', 'length', 'width']
print(df[dimension_columns].describe())

print("\nMake-wise Average Dimensions:")
make_wise_summary = df.groupby('make')[dimension_columns].mean().round(2)
print(make_wise_summary)
```
